```python
#path="eco590_Data_Analysis_Python_R\Lecture_6\ds"
wjp=pd.read_csv('wjp.csv')
ineq= pd.read_csv('ineq.csv')
```

```python
#mergeing wjp and ineq
wjp_ineq = pd.merge(wjp, ineq, left_on=['Country','year'],right_on=['country','year'],how='left')

wjp_ineq.head(5)
```

```python
#Filter out the dataset to show only data from the region 'Sub-Saharan Africa'
filter_region=wjp_ineq[wjp_ineq['Region']=="Sub-Saharan Africa"]
filter_region.head(5)
```

```python
#How many countries are in the region?
filter_region['Country'].nunique() #length func?
```

```python
#Calculate the average gini of the region.
filter_region[['gini']].head(5)

filter_region[['gini']].mean()
```
